```python
# fill the null values
wine_quality['total sulfur dioxide'].fillna(value=wine_quality['total sulfur dioxide'].mean(), inplace=True)
wine_quality['pH'].fillna(value=wine_quality['pH'].mean(), inplace=True)
wine_quality['quality'].fillna(value=wine_quality['quality'].mean(),inplace=True)

```

```python
# Separate the data label(quality)
X = wine_quality.drop('quality',axis = 1)
```

```python
print(X)
```

```python
#label binarization
Y = wine_quality['quality'].apply(lambda y_value:1 if y_value>=7 else 0)
```

```python
print(Y)
```

###### Train and test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
 X, Y, test_size=0.3, random_state=10)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

###### Model Training
###### Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier
RFC_model = RandomForestClassifier()
```

```python
RFC_model.fit(X_train,Y_train)
```

###### Accuracy Score

```python
#On test data

X_test_prediction = RFC_model.predict(X_test)
```

```python
from  sklearn.metrics import accuracy_score
test_data_accuracy= accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy:', test_data_accuracy)
```

##### Building Predictive Model

```python
Input_data = (11.8,0.26,0.52,1.8,0.071,6,10,0.9968,3.52,0.72,10.2)

## Changing the input data to a numpy array
Input_data_as_np_array = np.asarray(Input_data)

#reshape the data as we are predicting the label for only one instance
Input_data_reshaped = Input_data_as_np_array.reshape(1,-1)

prediction = RFC_model.predict(Input_data_reshaped)
print(prediction)


if (prediction[0] == 1):
    print('Its Good quality wine')
else:
    print('Its Bad quality Wine ')
```
